```python
# Rename columns for readability
combined_df = combined_df.rename(columns={"tripduration": "Trip Duration (Sec)",
                                                "starttime": "Start Time",
                                                "stoptime": "Stop Time",
                                                "start station id": "Start Station ID",
                                                "start station name": "Start Station Name",
                                                "start station latitude": "Start Station Latitude",
                                                "start station longitude": "Start Station Longitude",
                                                "end station id": "End Station ID",
                                                "end station name": "End Station Name",
                                                "end station latitude": "End Station Latitude",
                                                "end station longitude": "End Station Longitude",
                                                "bikeid": "Bike ID",
                                                "usertype": "User Type",
                                                "birth year": "User Birth Year",
                                                "gender": "User Gender"
                                                })
combined_df
```

```python
#remove unknown values from gender column 1 = male; 2 = female and 0 = unknown
unknown_gender = combined_df[combined_df['User Gender'] == 0].index
combined_df.drop(unknown_gender, inplace=True)
#count
combined_df.count()
```

```python
#create day column
combined_df['Day'] = pd.DatetimeIndex(combined_df['Start Time']).day
#create month column
combined_df['Month'] = pd.DatetimeIndex(combined_df['Start Time']).month
#create year column
combined_df['Year'] = pd.DatetimeIndex(combined_df['Start Time']).year
#create minute column for trip duration
combined_df['Trip Duration (Min)'] = pd.to_timedelta(combined_df['Trip Duration (Sec)'], unit= 's')

combined_df
```

```python
combined_df['User Gender'].replace([1],['Male'],inplace=True)
combined_df['User Gender'].replace([2],['Female'],inplace=True)
combined_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
combined_df.to_csv("bikedata.csv", index=False, header=True)
```
